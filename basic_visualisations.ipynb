```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
happiness = pd.read_csv('2019.csv', 'r', delimiter = ',')
happiness.head()
```

```python
happiness.info()
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 11,5
```

```python
happiness.plot(title = 'график счастья и ожидаемой продолжительности жизни', 
               x = 'Score',
               y = 'Healthy life expectancy',
               xlabel = 'счётчик счастья', 
               ylabel = 'ожидаемая продолжительность жизни')
# в менее счастливых регионах продолжительность жизни стабильно невысокая, с ростом счастья растёт и продолжительность жизни, при средних показателях счастья -
# - продолжительность нестабильна, при высоких - стабильно высокая. эти два значения зависят друг от друга. 	
```

```python
happiness.plot(kind = 'scatter',
               title = 'корреляция между щедростью и ВВП на человека',
               x = 'Generosity',
               y = 'GDP per capita',
               xlabel = 'щедрость',
               ylabel = 'ВВП на человека')

# при низких показателях ввп щедрость остаётся на среднем уровне, при этом при более высоком ввп на человека - щедрость скорее меньше, точек в правой части довольно мало,
# в целом, люди более склонны к жадности
```

```python
happiness['Perceptions of corruption'].plot(kind = 'box',
                        title = 'распределение представлений об уровне коррупции', grid = True)

# люди склонны верить, что уровень коррупции довольно низкий, либо резко противоположно - что крайне высок, редко считают, что средний
```

```python
top_20_regions = happiness['Freedom to make life choices'].head(20)
top_20_regions.plot(kind = 'bar',
                   title = 'топ 20 счастливых стран и их свобода выбора',
                   xlabel = 'страна в топе',
                   ylabel = 'свобода выбора')
plt.show()
last_20_regions = happiness['Freedom to make life choices'].tail(20)
last_20_regions.plot(kind = 'bar',
                   title = 'последние 20 стран по уровню счастья и их свобода выбора',
                   xlabel = 'страна в топе',
                   ylabel = 'свобода выбора')
# если у более статистически "счастливых" стран прослеживается стабильная высота свободы выбора, то в наименее счастливых странах этот показатель нестабилен и скачки между 
# значениями довольно большие. этот фактор не всегда добавляет счастья жителям, однако счастливые люди всегда обладают высокой свободой выбора
```
